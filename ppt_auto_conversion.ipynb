```python
from pptx import Presentation
prs = Presentation('/content/gdrive/MyDrive/2학기 프로젝트/초본/Standard_ppt.pptx')
```

```python
#첫 번째 슬라이드 레이아웃-논문 한글 제목, 논문 영어 제목, 이름, 소속 
add_slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(add_slide_layout)
# 슬라이드 내 개체 확인
shapes = slide.shapes
for shape in shapes:
  print(str(shape.placeholder_format.idx) + " : " + shape.name) 
```

```python
slide.placeholders[0].text = "논문 한글 제목 성공"
slide.placeholders[14].text = "학과 성공"
slide.placeholders[15].text = "이름 성공"
slide.placeholders[16].text = "논문 영어 제목 성공"
# 슬라이드 노트
notes_slide = slide.notes_slide
text_frame = notes_slide.notes_text_frame
text_frame.text = '권기현 바보'
```
